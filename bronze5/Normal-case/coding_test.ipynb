```
# no.1000 (A + B)
A, B = input('').split(' ')
print(int(A) + int(B))
```

```
# no.1001 (A - B)
A, B = input('').split(' ')
print(int(A) - int(B))
```

```
# no.1271 (엄청난부자2)
n, m = map(int, input('').split(' '))
quotient, remainder = divmod(n, m)
print(quotient)
print(remainder)
```

```
# no.1550 (16진수)
value = input()
value_16 = '0x{}'.format(value)
print(int(value_16, 16))
```
